```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = '/home/andjela/projects/humidity project/weatherHistory.csv'
DF = pd.read_csv(file_path)

# Convert 'Formatted Date' to datetime
DF['Formatted Date'] = pd.to_datetime(DF['Formatted Date'], format='%Y-%m-%d %H:%M:%S.%f %z')
```

```python
new_start_date = '2006-08-01 00:00:00.000 +0200'
new_end_date = '2006-08-09 23:00:00.000 +0200'
new_start_date = pd.to_datetime(new_start_date, format='%Y-%m-%d %H:%M:%S.%f %z')
new_end_date = pd.to_datetime(new_end_date, format='%Y-%m-%d %H:%M:%S.%f %z')
```

```python
filtered_df_new_period = DF[(DF['Formatted Date'] >= new_start_date) & (DF['Formatted Date'] <= new_end_date)]
```

```python
filtered_df_new_period.head()
```

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(x='Formatted Date', y='Humidity', data=filtered_df_new_period)
plt.title('Scatter Plot of Humidity from 2006-08-01 to 2006-08-09')
plt.xlabel('Date')
plt.ylabel('Humidity')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
